```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
path = "Plots and Results/LBM/"
```

```python
s20 = path+"20%.txt"
s50 = path+"50%.txt"
s80 = path+"80%.txt"
s100 = path+"100%.txt"
```
